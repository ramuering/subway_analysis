```python
import folium
import pandas as pd
data=pd.read_csv('sub_con_fin.csv', encoding='euc-kr')

# 지도 객체 생성
m = folium.Map(location=[37.508725,126.891295], zoom_start=10)

# 데이터를 지도에 추가
for index,row in data.iterrows():
    lat = row['lat']
    lon = row['lng']
    reservoir_ratio = row['con']
    # 마커 크기 지정
    if reservoir_ratio >150:
       radius=20
       fill_color='red'
       
    elif reservoir_ratio >100:
       radius=13
       fill_color='orange'
       
    elif reservoir_ratio >50:
       radius=10
       fill_color='green'
       
    else:
       radius=7
       fill_color='blue'
       
      
      
    # 팝업 텍스트 설정
    popup_text = f"""
        <b>행정동: </b>{row['행정동']}<br><b>혼잡도: </b>{row['con']}
    """
    # 팝업 텍스트 크기 조정
    popup = folium.Popup(popup_text, max_width=500)
    # 마커 생성 및 지도에 추가
    marker = folium.CircleMarker(
        location=[lat, lon],
        radius=radius,
        color='black',
        fill_color=fill_color,
        fill_opacity=0.5,
        popup=popup
    )
    marker.add_to(m)

m
```
