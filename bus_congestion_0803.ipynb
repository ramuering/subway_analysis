```python
import folium
import pandas as pd
data=pd.read_csv('bus_congestion_0803.csv', encoding='euc-kr')

# 지도 객체 생성
m = folium.Map(location=[37.508725,126.891295], zoom_start=10)

# 데이터를 지도에 추가
for index,row in data.iterrows():
    lat1 = row['lon']
    lon = row['lat']
    reservoir_ratio = row['승차인원']
    # 마커 크기 지정
    if reservoir_ratio >80000:
       radius=50
       opac=1
    elif reservoir_ratio >40000:
       radius=35
       opac=0.8
    elif reservoir_ratio >20000:
       radius=20
       opac=0.6
    elif reservoir_ratio > 10000:
       radius=11
       opac=0.4
    elif reservoir_ratio>5000:
       radius=6
       opac=0.2
    else:
       radius=3
       opac=0.1
      
      
    # 팝업 텍스트 설정
    popup_text = f"""
        <b>행정동: </b>{row['행정동']}<br><b>혼잡도: </b>{row['승차인원']}
    """
    # 팝업 텍스트 크기 조정
    popup = folium.Popup(popup_text, max_width=500)
    # 마커 생성 및 지도에 추가
    marker = folium.CircleMarker(
        location=[lat1, lon],
        radius=radius,
        color='black',
        fill_color='chocolate',
        fill_opacity=opac,
        popup=popup
    )
    marker.add_to(m)

m
```
